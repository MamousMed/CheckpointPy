```python
import matplotlib.pyplot as plt
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)
```

```python
'''The correlation map function processes the information in order to show us how strong is the relationship between every two features which allows us to understand the impact of each feature on the other one. This way we can extract the most important features and drop the least important ones. For example, in this plot we can see that relationship between Pclass and Fare is close to -1 and the one between SibSp and Parch is somewhat closer to 1 however the relationship between Pclass and SibSp is closer to 0. From this data, we can conclude that there is a strong correlation between Pclass/Fare and SibSp/Parch and a weak relationship between Pclass/Parch'''
```
